```python
canc_per=df['is_canceled'].value_counts(normalize=True)
canc_per*100

plt.figure(figsize=(5,4))
plt.title("Reservation status")
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts())
plt.show()
```

```python
plt.figure(figsize=(8,4))
ax1=sns.countplot(x='hotel',hue="is_canceled",data=df)
plt.title("Reservation cancelled")
plt.xlabel("hotel")
plt.ylabel("No. of reservations")
plt.legend(['not canceled','canceled'])
plt.show()
```

```python
resort_hotel=df[df['hotel']=="Resort Hotel"]
resort_hotel["is_canceled"].value_counts(normalize=True)*100
plt.figure(figsize(8,4))
sns.barplot(x='resort_hotel')
```

```python
city_hotel=df[df['hotel']=="City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True)*100
```

```python
resort_hotel=resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel=city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(15,6))
plt.title("Avg Daily rate in resort hotel and city hotel")
plt.plot(resort_hotel.index,resort_hotel['adr'])
plt.plot(city_hotel.index,city_hotel['adr'])
```

```python
df['month']=df['reservation_status_date'].dt.month
plt.figure(figsize=(15,6))
ax1=sns.countplot(x='month',hue='is_canceled',data=df)
plt.title("Reservation status per month")
plt.xlabel("Months")
plt.ylabel("Reservation")
plt.legend(['not canceled','canceled'])
```

```python
city_hotel=df[df['hotel']=="City Hotel"]
city_hotel["is_canceled"].value_counts(normalize=True)*100 
```

```python
plt.figure(figsize=(15,8))
plt.title("ADR PER MONTH")
sns.barplot('month','adr',data=df[df['is_canceled']==1].groupby('month')[['adr']].sum().reset_index())
plt.xlabel("Month")
plt.ylabel("ADR")
plt.show()
```

```python
cancelled_data=df[df["is_canceled"]==1]
top_10_countries=cancelled_data["country"].value_counts()[:10]
plt.figure(figsize=(10,8))
plt.pie(top_10_countries, autopct="%.2f" ,labels=top_10_countries.index)
plt.title("Top 10 countries reservation cancellation ", fontsize=25)
plt.show()
```

```python
calc=df['market_segment'].value_counts(normalize=True)*100
plt.figure(figsize=(20,12))
plt.pie(calc,autopct='%.2f',labels=calc.index)
plt.title("BOOKINGS DONE THROUGH ",fontsize=20)
plt.show()
```

```python
cancelled_booking=cancelled_data['market_segment'].value_counts(normalize=True)*100
plt.figure(figsize=(20,12))
plt.pie(cancelled_booking,autopct='%.2f',labels=cancelled_booking.index)
plt.title("CANCELLATION DONE THROUGH ",fontsize=20)
plt.show()
```
